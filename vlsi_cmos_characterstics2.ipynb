# N-mos IV curve 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
beta=0.000240
Vt=0.7
Vds=np.linspace(0,5,num=100)
Vgs=np.arange(1,6)
```

```python
plt.figure(figsize=(15,8))
for j in range (len(Vgs)):
    Ids=[]
    for i in range(len(Vds)):
        if Vgs[j]<Vt:
            Ids1=0
        elif (Vgs[j]>Vt) & (Vds[i]<=(Vgs[j]-Vt)):
            Ids1=(beta*((Vgs[j]-Vt)-Vds[i]/2))*Vds[i]
        elif (Vgs[j]>Vt) & (Vds[i]>(Vgs[j]-Vt)):
            Ids1=(0.5*beta)*(Vgs[j]-Vt)**2
        Ids.append(Ids1*1000)
    plt.plot(Vds,Ids,linewidth=3,label='Vgs ='+str(Vgs[j]))
plt.title('I-V curve of NMOS')
plt.xlabel('Vds (volts)')
plt.ylabel('Ids (mA)')
plt.ylim([0, 3])
plt.xlim([0, 5])
plt.legend(loc= 'upper right', shadow=True)
plt.grid()
plt.show()
```

# P-mos IV curve

```python
Vt2=-0.7
Vds2=np.linspace(-5,0,num=100)
Vgs2=np.arange(-5,0)
```

```python
plt.figure(figsize=(15,8))
for j in range (len(Vgs2)):
    Ids=[]
    for i in range(len(Vds2)):
        if Vgs2[j]>Vt2:
            Ids2=0
        elif (Vgs2[j]<Vt2) & (Vds2[i]>=(Vgs2[j]-Vt2)):
            Ids2=(-beta*((Vgs2[j]-Vt2)-Vds2[i]/2))*Vds2[i]
        elif (Vgs2[j]<Vt2) & (Vds2[i]<(Vgs2[j]-Vt2)):
            Ids2=(-0.5*beta)*(Vgs2[j]-Vt2)**2
        Ids.append(Ids2*1000)
    plt.plot(Vds2,Ids,linewidth=3,label='Vgs =' +str(Vgs[j]))
plt.title('I-V curve of PMOS')
plt.xlabel('Vds (volts)')
plt.ylabel('Ids (mA)')
plt.ylim([-3,0])
plt.xlim([-5, 0])
plt.legend(loc= 'lower left', shadow=True)
plt.grid()
plt.show()
```

# CMOS load line Analysis

```python
plt.figure(figsize=(15,10))
for j in range (len(Vgs)):
    Ids=[]
    for i in range(len(Vds)):
        if Vgs[j]<Vt:
            Ids1=0
        elif (Vgs[j]>Vt) & (Vds[i]<=(Vgs[j]-Vt)):
            Ids1=(beta*((Vgs[j]-Vt)-Vds[i]/2))*Vds[i]
        elif (Vgs[j]>Vt) & (Vds[i]>(Vgs[j]-Vt)):
            Ids1=(0.5*beta)*(Vgs[j]-Vt)**2
        Ids.append(Ids1*1000)
    plt.plot(Vds,Ids,linewidth=3)
    Vds2=np.flip(Vds)
    plt.plot(Vds2,Ids,linewidth=3)
plt.title('Load line analysis of CMOS inverter')
plt.xlabel('Vout (volts)')
plt.ylabel('Ids (mA)')
plt.ylim([0, 3])
plt.xlim([0, 5])
plt.grid()
plt.show()
```
